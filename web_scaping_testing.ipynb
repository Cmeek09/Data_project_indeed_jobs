```python
## gets job searches

import requests
import pandas as pd

url = "https://indeed12.p.rapidapi.com/jobs/search"

query_params = {
    "query": "software developer",
    "location": "us",
    "page_id": 1  # Start with the first page
}

headers = {
    "X-RapidAPI-Key": "3768c978a8msh0edeb2fb8b71f43p1d1608jsn115418a7ae39",
    "X-RapidAPI-Host": "indeed12.p.rapidapi.com"
}

all_data = []  # To store all job listings
max_pages = 40  # Maximum number of pages to fetch

for _ in range(max_pages):
    response = requests.get(url, params=query_params, headers=headers)
    data = response.json()

    # Append current page's job listings to the overall list
    all_data.extend(data["hits"])

    # Check if there are more pages available
    if "next_page_id" not in data:
        break

    # Update the query parameter for the next page
    query_params["page_id"] = data["next_page_id"]

df = pd.DataFrame(all_data)

# Save DataFrame to Excel
df.to_excel("job_listings.xlsx", index=False)

print("DataFrame saved to 'job_listings.xlsx'")
```
